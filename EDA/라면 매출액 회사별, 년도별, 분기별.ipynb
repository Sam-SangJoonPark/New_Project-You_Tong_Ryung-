```python
def process_1(ramen,quarter) :
    ramen = ramen[ramen['분기'] ==quarter]
    return ramen

ramen_1 = process_1(ramen,1)
ramen_1
```

```python
def process_2(ramen_1,year):
    ramen_201x_n = ramen_1[ramen_1['년도']==year]
    return ramen_201x_n
ramen_201x_n = process_2(ramen_1,2018)
ramen_201x_n
```

```python
def process_3(ramen_201x_n, comp) :
    ramen_201x_n_comp = ramen_201x_n[ramen_201x_n['제조사'] == comp]
    return ramen_201x_n_comp

ramen_201x_n_comp = process_3(ramen_201x_n, '농심')
ramen_201x_n_comp
```

```python
def process_4(ramen_201x_n_comp) :
    a = ramen_201x_n_comp[['매출액(백만원)']].sum()
    return a 

process_4(ramen_201x_n_comp)
```

```python
def process_one(ramen,year,quarter,comp) :
    a = process_1(ramen,quarter)
    b = process_2(a, year)
    c = process_3(b, comp)
    d = process_4(c)
    return d
    
```

```python
ottugi =  []
samyang = []
nongshim = []
for year in [2018,2019] :
    for quarter in [1,2,3,4] :
        for company in ['오뚜기', '농심', '삼양'] :
            print('*****start******',(year,quarter, company),process_one(ramen, year,quarter, company),'*****end****')
            if company == '오뚜기' :
                ottugi.append(process_one(ramen, year,quarter, company))
            if company == '삼양' :
                samyang.append(process_one(ramen, year,quarter, company))
            if company == '농심' :
                nongshim.append(process_one(ramen, year,quarter, company))
            
```
