```python
# https://yganalyst.github.io/data_handling/Pd_12/
data = pd.concat([data1, data2, data3, data4,data5,data6,data7,data8,data9,data10,data11], axis = 0)  #행방향 연결, 데이터프레임
data
```

```python
for i in range(len(data['0'].values)) :
    if len(data['0'].values[i]) < 10 :
        data['0'].values[i] = '2020년 ' + data['0'].values[i]
    
```

```python
data
```

```python
for i in range(len(data['0'].values)) :
    data['0'].values[i] = data['0'].values[i].replace('년','-')
    data['0'].values[i] = data['0'].values[i].replace('월','-')
    data['0'].values[i] = data['0'].values[i].replace('일','')
    data['0'].values[i] = data['0'].values[i].replace(' ','')
 
```

```python
for i in range(len(data['0'].values)) :
    a = data['0'].values[i].split('-')
    if len(a[1]) == 1 :
        a[1] = '0' + a[1]
    data['0'].values[i] = a[0]+ '-' + a[1] + '-' + a[2]
    
```

```python
def my_process(data) :
    # data 날짜를 리스트로
    
    list_data = []
    for i in data['0'] :
        list_data.append(i)

    # data에서 2019년 빼기
    for i in range(len(list_data)) :
        list_data[i] = list_data[i][:9]
        
  
        
    return list_data

list_data = my_process(data)
```

```python
list_data
```

```python
def my_common(list_data) : 
    from collections import Counter
    count = Counter(list_data)

    count = sorted(count.items())
    return count

count = my_common(list_data)


date_list = []
for (i,a) in count :
    date_list.append(i)
```

```python
def sort_date(date_list) :
    for i in range(len(date_list)) :
        if str(date_list[i])[0] != '0' :
            date_list[i] = str(date_list[i])[:2] + '월 ' + str(date_list[i])[2:4] + '일'
        else :
            date_list[i] = str(date_list[i])[1] + '월 ' + str(date_list[i])[2:4] + '일'

        if str(date_list[i])[-3] == '0' :
            date_list[i] = str(date_list[i])[:-3] + str(date_list[i])[-2:]
    return date_list

date_list = sort_date(date_list)
```

```python
data_num = {}
for (a,b) in count :
    data_num[a] = b
data_num
```

```python
list_data = []
data_num_1 = []
for (a,b) in data_num.items() :
    list_data.append(a)
    data_num_1.append(b)
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib as mpl


days=list_data 
days.sort()
xtic=np.arange(0,len(list_data),step=30)   
xdays=[days[i] for i in xtic] 
plt.xticks(xtic,xdays,rotation=-30)
plt.title("green_muchim mentioned frequency in instagram")   # 이름지정
plt.xlabel('Date', size = 20)
plt.ylabel('Frequency', size = 15)
plt.axvline(x=117,label="오뚜기 미역초 비빔면 출시", color='r')

plt.plot(list_data,data_num_1)
plt.show()
```
