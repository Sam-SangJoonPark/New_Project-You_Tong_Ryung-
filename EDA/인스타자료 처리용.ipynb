```python
import pandas as pd
import numpy as np
import matplotlib as mlp
import matplotlib.pyplot as plt
```

```python
# data 1~ n 까지 !
data1 = pd.read_csv("C:/Users/ASIA_07/유통프로젝트/truffle_insta{2655}.csv")
data2 = pd.read_csv("C:/Users/ASIA_07/유통프로젝트/truffle_insta{5278}.csv")
```

```python
# https://yganalyst.github.io/data_handling/Pd_12/
data = pd.concat([data1, data2], axis = 0)  #행방향 연결, 데이터프레임
data
```

```python
def my_process(data) :
        # data 날짜를 리스트로
    list_data = []
    for i in data['0'] :
        list_data.append(i)

    # data에서 2019년 빼기
    for i in range(len(list_data)) :
        list_data[i] = list_data[i][6:]
        
    return list_data
```

```python
def my_common(list_data) :
    from collections import Counter 
    count = Counter(list_data) 
    common_tag = count.most_common()
    common_tag.sort()
    common_tag
    data_num = []
    for (i,j) in (common_tag) :
        data_num.append(j)
        
    return data_num
```

```python
# 10월, 11일이 먼저나오므로 다시 정렬이 필요함!
def my_list(list_data) :
    list_data = list(set(list_data))
    list_data.sort()
    list_data
    return list_data
```

```python

def My_Data(data) :
    list_data = my_process(data)
    data_num = my_common(list_data)
    list_data = my_list(list_data)
    import matplotlib.pyplot as plt
    import matplotlib.font_manager as fm

    path = 'C:/Users/ASIA_07/Downloads/NanumFontSetup_TTF_ALL/NanumBarunGothic.ttf'
    fontprop = fm.FontProperties(fname=path, size=18)

    
    plt.title('트러플 언급', fontproperties=fontprop)
    plt.ylabel('빈도', fontproperties=fontprop)
    plt.plot(list_data,data_num)


    


    
    days=my_list(data['0'].values[:])   
    xtic=np.arange(0,len(my_list(my_process(data))),step=10)   
    xdays=[days[i] for i in xtic] 
    plt.xticks(xtic,xdays,rotation=90)
    
    plt.show()
        
    return plt.show()
```
