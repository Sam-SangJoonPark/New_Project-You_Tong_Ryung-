```python
""" 0  - 가전/가구 . 1 - 가정생활/서비스 . 2- 교육/학원 . 3 - 미용
    4 - 스포츠/문화/레저 .  5- 여행/교통  6. 요식/유흥 7. 유통
    8 - 음/식료품 . 9 - 의료 . 10 -자동차. 11 - 주유 . 12- 패션/잡화 """
    
    
def make_graph(num) :
    data_make = data[num::13]
    xtic = np.arange(0,len(data_make),step=30) 
    days = data_make['이용일자'].values[:]
    xdays = [days[i] for i in xtic]
    xtic = xtic*13
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)
    s1 = data_make["카드결제건수(천건)"]
    ax1 = s1.plot()
    plt.axvline(x=3440, label="중국 첫 코로나", color ='red')     # 빨간선은 코로나 시작 시점
    plt.title("the graph")
    plt.xticks(xtic,xdays, rotation = 90)
    plt.show()
```

```python
make_graph(5)
```
