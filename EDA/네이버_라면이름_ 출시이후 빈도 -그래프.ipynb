```python
import pandas as pd
import numpy as np
import matplotlib as mlp
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('C:/Users/ASIA_07/유통프로젝트/1.갈릭버터볶음면출시 이후.csv')
data =  data[['date','frequency']]
data
```

```python
for i in range(len(data)) :
    data['date'][i] = str(data['date'][i])[0:4] + '-' + str(data['date'][i])[4:6] + '-' + str(data['date'][i])[6:]
```

```python
data
```

```python
date_data =  list(data['date'].values)
freq_data =  list(data['frequency'].values)
```

```python
freq_data
```

```python
data_num = {}
for i in range(len(date_data)) :
    data_num[date_data[i]] = freq_data[i]

data_num
```

```python
list_data = []
data_num_1 = []
for (a,b) in data_num.items() :
    list_data.append(a)
    data_num_1.append(b)
```

```python
freq_data
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib as mpl


days=list_data 
days.sort()
xtic=np.arange(0,len(list_data),step=5)   
xdays=[days[i] for i in xtic] 
plt.xticks(xtic,xdays,rotation=-30)
plt.title("after ramyun release")   # 이름지정
plt.xlabel('Date', size = 20)
plt.ylabel('Frequency', size = 15)
plt.axvline(x=0,label="이 라면의 출시일", color='r')

plt.plot(list_data,data_num_1)
plt.show()
```
